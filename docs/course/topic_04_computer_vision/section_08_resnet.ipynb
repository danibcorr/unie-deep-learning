### La Degradacin del Gradiente

Antes de la introduccin de ResNet, predominaba la intuicin de que aumentar el nmero de capas conduca de forma casi automtica a una mayor precisin. Sin embargo, los experimentos empricos mostraron que, a partir de un umbral relativamente bajo en torno a veinte o treinta capas, el rendimiento comenzaba a deteriorarse de manera significativa.

Este fenmeno no se deba al *overfitting*, sino a un problema estructural ms profundo conocido como **degradacin**. En redes muy profundas, la seal de error que se propaga hacia atrs durante el proceso de entrenamiento tiende a debilitarse progresivamente. Como resultado, las capas ms cercanas a la entrada reciben gradientes muy pequeos o inestables, lo que dificulta o impide el aprendizaje efectivo. En este contexto, aadir ms capas no solo dejaba de ser beneficioso, sino que empeoraba la capacidad de la red para representar la funcin deseada.

### Conexiones Residuales

La innovacin central de ResNet consiste en la introduccin de las **conexiones de atajo**, tambin denominadas *skip connections*, que dan lugar al denominado **bloque residual**. En lugar de obligar a cada conjunto de capas a aprender directamente una transformacin completa, la arquitectura permite que dichas capas aprendan nicamente la **diferencia**, o residuo, entre la entrada y la salida.

Desde un punto de vista funcional, si una capa recibe una entrada \(x\), una red convencional aprende una transformacin \(F(x)\) y produce como salida \(F(x)\). En ResNet, la salida se define como \(F(x) + x\). Esta simple suma introduce una ruta alternativa por la que la informacin puede fluir sin modificaciones, creando una especie de autopista que atraviesa la red.

Este mecanismo tiene consecuencias profundas para el entrenamiento. Si una capa o bloque no resulta til, el modelo puede aproximar el residuo a cero, permitiendo que la seal original se propague sin alteraciones. Adems, durante la retropropagacin, el gradiente puede atravesar estas conexiones directas sin atenuarse, lo que estabiliza el aprendizaje incluso en redes extremadamente profundas.

### Variantes Arquitectnicas y Diseo Interno

La familia ResNet incluye mltiples variantes que se diferencian principalmente por su profundidad, entre las que destacan ResNet-18, ResNet-34, ResNet-50, ResNet-101 y ResNet-152. A partir de cincuenta capas, la arquitectura adopta un diseo optimizado conocido como **bloque bottleneck** o cuello de botella, cuyo objetivo es reducir el coste computacional sin sacrificar capacidad representacional.

Este bloque se compone de tres convoluciones consecutivas. La primera es una convolucin **1  1** que reduce la dimensionalidad del espacio de caractersticas, la segunda es una convolucin **3  3** que realiza el procesamiento principal, y la tercera es otra convolucin **1  1** que restaura la dimensionalidad original. Esta estrategia permite construir redes muy profundas manteniendo un nmero de parmetros relativamente contenido.

La comparacin entre LeNet-5, VGG-16 y ResNet-50 pone de manifiesto la evolucin conceptual del aprendizaje profundo. LeNet-5, presentado en 1998, cuenta con siete capas y unos sesenta mil parmetros, y su principal innovacin fue la introduccin de la convolucin como mecanismo para explotar la estructura espacial de las imgenes. ResNet-50, introducida en 2015, alcanza cincuenta capas con tan solo unos veinticinco millones de parmetros gracias al uso sistemtico de conexiones residuales, logrando una combinacin notable de profundidad, estabilidad y eficiencia.

Por estas razones, ResNet no solo resolvi un problema tcnico concreto, sino que redefini la forma de concebir la profundidad en las redes neuronales convolucionales.
